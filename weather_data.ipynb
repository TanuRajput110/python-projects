```python
#5.rename the column 'weather' to 'Weather Condition'
df.rename(columns={'Weather':'weather Condition'},inplace=True)
df.head()
```

```python
#6.What is the mean 'Visibility
df['Visibility_km'].mean()
```

```python
#7.What is the Standard deviation of 'Pressure' in this data
df['Press_kPa'].std()
```

```python
#8.what is the variance of 'relative humanity' in this data
df['Rel Hum_%'].var()
```

```python
#9.Find all the instance when 'Snow' was recorded
df[df['weather Condition'].str.contains('Snow')].head()
```

```python
#10 find all the instance when 'Wind Speed is above 24' and 'Visibility is 25'
df[(df['Wind Speed_km/h']>24) & (df['Visibility_km'] ==25)]
```

```python
#11.What is the mean value of each column against each 'Weather Condition'?
df.groupby('weather Condition').mean()
```

```python
#12.What is The minimum and maximum value of each column against each 'Weather Condition'?
df.groupby('weather Condition').min()
```

```python
#13.Show all the records where Weather condition is Fog.
df[df['weather Condition']=='Fog']
```

```python
#14.Find all instance when 'weather is clear' or  'Visibility is above 40'.
df[(df['weather Condition']=='Fog') | (df['Visibility_km'] > 40)]
```

```python
#15.Find all instance
#a.'Weather is clear' and 'Relative Humanity' is greater than 50
#or
#b.'Visibility is above 40

df[(df['weather Condition']=='Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
```
